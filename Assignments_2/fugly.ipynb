```python
import numpy as np
import pandas as pd
```

```python
X_train = pd.read_csv('training_set.csv', header=None, names=['x','y','t'])
X_val = pd.read_csv('validation_set.csv', header=None, names=['x','y','t'])

x = list(X_train['x'])
xn = (x - np.mean(x))/np.std(x)
y = list(X_train['y'])
yn = (y - np.mean(y))/np.std(y)
t = list(X_train['t'])


xv = list(X_val['x'])
yv = list(X_val['y'])
xvn = (xv - np.mean(xv))/np.std(xv)
yvn = (yv - np.mean(yv))/np.mean(yv)
tv = list(X_val['t'])

XT = np.zeros((10000,3))
XT[:,0] = xn
XT[:,1] = yn
XT[:,2] = t

XV = np.zeros((5000, 3))
XV[:,0] = xvn
XV[:,1] = yvn
XV[:,2] = tv

```

```python
# Network time
M = 20
p_train = 10000
p_val = 5000
W = np.random.normal(0,1,(M,))
w = np.random.normal(0,1,(M,2))
nu = 0.005

theta = np.zeros((M,))
T = 0

epochs = 1000
C_vec = np.zeros((epochs))

run = 1
iter_index = 0

while run:
    
    O_vec = np.zeros((p_train,))
    for i_ in range(p_train): # One epoch = p_train long
        idx = np.random.choice(p_train) # choose random patter
        
        t = XT[idx,2]
        x = XT[idx,:2]
        
        b = w @ x - theta
        V = np.tanh(b)
        
        B = np.dot(W, V)
        O = np.tanh(B)
        O_vec[i_] = O
        
        # Gradient decsent
        d_B = (t - O) * (1/(np.cosh(B))) ** 2
        d_b = np.zeros((M,))
        
        for l in range(M):
            d_b[l] = d_B * W[l] * (1/(np.cosh(b[l])))**2
        
        dW = nu*d_B*V
        W += dW
        
        dw = nu*np.outer(d_b,x)
        w += dw
        
        dT = -nu*d_B
        T += dT
        dtheta = -nu*d_b
        theta += dtheta
        
    # Validation time
    O_val = np.zeros((p_val,1))
    
    for v in range(p_val):
        t = XV[v,2]
        x = XV[v,:2]
        
        b = w @ x - theta
        V = np.tanh(b)
        B = np.dot(W, V)
        O = np.tanh(B)
        O_val[v] = O
        
    SO = np.sign(O_val)
    SO[SO==0] = 1
    if iter_index % 100 == 0:
        print(iter_index)
    C = np.sum(np.abs(SO - XV[:,2]))/(2*p_val)
    C_vec[iter_index] = C
    if C <0.116:
        run = 0
    iter_index += 1

print(f'C={C} found after during iteration: {iter_index}')
```
